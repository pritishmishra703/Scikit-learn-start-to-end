# MNIST dataset with Logistic Regression

```python
from sklearn.datasets import load_digits
```

```python
mnist = load_digits()
```

```python
X = mnist.data
```

```python
y = mnist.target
```

```python
mnist.target_names
```

```python
import matplotlib.pyplot as plt

plt.gray()
plt.matshow(mnist.images[23])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

X = StandardScaler().fit_transform(X)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
from sklearn.linear_model import LogisticRegression

reg = LogisticRegression()
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
reg.score(X_test, y_test)
```

```python
from PCM.PCM import plot_confusion_matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
plot_confusion_matrix(cm, mnist.target_names)
```

```python
index = 0
misclassified = []
for i, j in zip(y_test, y_pred):
    if i != j:
        misclassified.append(index)
    index += 1
```

```python
misclassified
```

```python
for i in misclassified:
    plt.gray()
    plt.matshow(X_test[i].reshape(8, 8))
    plt.title(f"Actual: {y_test[i]}, Predicted: {y_pred[i]}")
    plt.show()
```
